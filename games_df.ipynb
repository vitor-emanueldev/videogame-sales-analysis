# Imports e carregamento dos dados

```python
import pandas as pd

df_games = pd.read_csv("games.csv")
display(df_games.head())
```

# Inspeção inicial dos dados
Verificando inconsistências na coluna `platform` — identificamos jogos modernos cadastrados no Game Boy com datas e títulos inválidas. Além disso, o dataset possui muitas colunas que não serão utilizadas nesta análise e serão removidas na etapa de limpeza.

```python
# Inspeção
display(df_games.info())
display(df_games.describe())
display(df_games.columns)

df_games[df_games['platform'] == 'Game Boy']
df_games[df_games['platform'] == 'Game Boy'][['title', 'platform', 'year']].sort_values('year', ascending=False).head(20)
```

# Limpeza dos dados
Remoção de colunas desnecessárias para esta análise e exclusão de registros inválidos — jogos cadastrados incorretamente na plataforma Game Boy, com títulos e datas incompatíveis com o hardware.

```python
df_gamesAnalise = df_games.drop(columns=["platform_type", 'publisher_region', 'publisher_tier', 'esrb_rating', 'critic_review_count', 'user_review_count', 'is_sequel', 'online_multiplayer', 'dlc_released', 'microtransactions', 'loot_boxes',
       'game_pass_available', 'vr_support', 'goty_nominated', 'goty_won',
       'how_long_to_beat_main_hrs', 'how_long_to_beat_completionist_hrs', 'platform_generation'])

display(df_gamesAnalise.info())
df_gamesAnalise = df_gamesAnalise.drop_duplicates(subset=["game_id"])

df_gamesAnalise[df_gamesAnalise['platform'] == 'Game Boy']
df_gamesAnalise[df_gamesAnalise['platform'] == 'Game Boy'][['title', 'platform', 'year']].sort_values('year', ascending=False).head(20)

df_gamesAnalise_limpo = df_gamesAnalise[df_gamesAnalise['platform'] != 'Game Boy']
df_gamesAnalise_limpo = df_gamesAnalise_limpo.reset_index(drop=True)

display(df_gamesAnalise_limpo)
```

```python
publish_mais_lucrativa = df_gamesAnalise_limpo[["publisher","estimated_revenue_million_usd"]].groupby("publisher").sum()
publish_mais_lucrativa = publish_mais_lucrativa.sort_values(by="estimated_revenue_million_usd", ascending=False)
publish_mais_lucrativa = publish_mais_lucrativa.reset_index()

display(publish_mais_lucrativa)
```

```python
gener_popular = df_gamesAnalise_limpo[["genre", "na_sales_million", "eu_sales_million", "jp_sales_million"]].groupby("genre").mean()

display(gener_popular)
```
